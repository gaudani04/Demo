### 06) WAP to demonstrate the use of Scatter Plot.

```python
import random
import matplotlib.pyplot as plt
random.seed(10)
x = [random.random() for i in range(0,100)]
y = [random.random() for i in range(0,100)]
z = [random.random() for i in range(0,100)]


plt.scatter(x,y,c=z,cmap='plasma')
plt.colorbar()
```

### 07) WAP to demonstrate the use of Histogram.

```python
import random
import matplotlib.pyplot as plt
random.seed(10)
age = [random.randint(1,50) for i in range(100)]
plt.hist(age,bins=5,histtype = 'step')
plt.xticks([1,11,21,31,41,51])
```

### 08) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
months = ['Jan','Feb','Mar','Apr','May','Jun','jul','Aug','sep','oct','nov','desc']
visitors = [11000,9000,5700,5400,12000,43000,9000,3000,2000,4000,9500,1400]
bars  = plt.bar(months,visitors,color='skyblue')
plt.xlabel("Months")
plt.ylabel("Visitors")
plt.title("Monthly Visitors")
for i in bars:
    yc = i.get_height()
    plt.text(i.get_x(),yc+50,f"{yc}")
plt.show()
```

### 09) WAP create a Scatter Plot with several colors in Matplotlib?
